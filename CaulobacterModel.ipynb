```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
time_v1 = [0,14.5,24.75,40.6,46.25,49.25]
```

```python
nomellini_rep1_v1 = [0,0,0.138,0.944,1.272,1.168]
nomellini_rep2_v1 = [0,0.013,0.32,1.152,1.304,1.32]
nomellini_combined_v1 = []
yerr_nomellini = []
for i,j in zip( nomellini_rep1_v1, nomellini_rep2_v1 ):
    nomellini_combined_v1 += [ (i + j)/2 ]
    arr = np.array( [i,j] )
    yerr_nomellini += [np.std(arr)]
```

```python
restreaked_rep1_v1 = [0,0.015,0.149,1.344,1.496,1.512]
restreaked_rep2_v1 = [0,0, 0.249,1.32,1.32,1.352]
restreaked_combined_v1 = []
yerr_restreaked = []
for i,j in zip( restreaked_rep1_v1, restreaked_rep2_v1 ):
    restreaked_combined_v1 += [ (i+j)/2 ]
    arr = np.array( [i,j] )
    yerr_restreaked += [np.std(arr)]
```

```python
plt.plot( time_v1, nomellini_combined_v1,marker='o' )
plt.xlabel('Time (in hours)')
plt.ylabel('Optical Density')
plt.errorbar( time_v1, nomellini_combined_v1, yerr=yerr_nomellini)
plt.title('C. crescentus ( Dr. Nomellini ) Growth over time')
plt.show()
```

```python
plt.plot( time_v1, restreaked_combined_v1,marker='o' )
plt.xlabel('Time (in hours)')
plt.ylabel('Optical Density')
plt.errorbar(time_v1, restreaked_combined_v1,yerr=yerr_restreaked)
plt.title('C. crescentus ( Restreaked Mar 7) Growth over time')
plt.show()
```

## Combined replicates of Caulobacter Crescentus

```python
yerr_combined = []
combined = []
for i,j,k,l in zip( nomellini_rep1_v1,nomellini_rep2_v1, restreaked_rep1_v1, restreaked_rep2_v1 ):
    combined += [ (i+j+k+l)/2 ]
    arr = np.array( [i,j,k,l] )
    yerr_combined += [np.std(arr)]
```

```python
plt.plot( time_v1, combined, marker='o' )
plt.xlabel('Time (in hours)')
plt.ylabel('Optical Density')
plt.errorbar( time_v1, combined, yerr=yerr_combined)
plt.title('C. rescentus Growth over time')
plt.show()
```
